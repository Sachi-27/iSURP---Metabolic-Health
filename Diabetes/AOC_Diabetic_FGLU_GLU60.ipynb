```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math

data = pd.read_excel('diabetic_glucose.xlsx')
data[data['GLU15'] == 0]
data.drop(data.index[104], inplace = True)
data.reset_index(drop=True, inplace=True)
data.describe()
```

```python
sns.heatmap(data.corr().round(2), annot=True)
```

We are now going to examine whether we can do any better given FGLU and GLU60 at the start

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

### Approach 1

```python
fglu_values = np.array(data['FGLU'].values)
glu15_values = np.array(data['GLU15'].values)
glu30_values = np.array(data['GLU30'].values)
glu60_values = np.array(data['GLU60'].values)
glu90_values = np.array(data['GLU90'].values)
glu120_values = np.array(data['GLU120'].values)
```

```python
params = pd.read_csv('Approach1.csv')
params
```

```python
def glu15_from_fglu(x):
    return x*params['Slope'].iloc[0] + params['Intercept'].iloc[0]

def glu30_from_glu15(x):
    return x*params['Slope'].iloc[1] + params['Intercept'].iloc[1]
def glu15_from_glu30(x):
    return (x-params['Intercept'].iloc[1])/params['Slope'].iloc[1]

def glu30_from_glu60(x):
    return (x-params['Intercept'].iloc[2])/params['Slope'].iloc[2]
def glu60_from_glu30(x):
    return (x-params['Intercept'].iloc[2])/params['Slope'].iloc[2]

def glu90_from_glu60(x):
    return x*params['Slope'].iloc[3] + params['Intercept'].iloc[3]

def glu120_from_glu90(x):
    return x*params['Slope'].iloc[4] + params['Intercept'].iloc[4]
def glu90_from_glu120(x):
    return (x-params['Intercept'].iloc[4])/params['Slope'].iloc[4]
```

#### True AOC values

```python
AOC = []

x = [0, 15, 30, 60, 90, 120]
for i in range(len(data)):
    AOC.append(np.trapz(data.loc[i].values, x))
```

### Stage 1 AOC Approximation

AOC only with points FGLU and GLU120

```python
Approx_AOC1 = []

for i in range(len(data)):
    Approx_AOC1.append(2*np.trapz([data.loc[i].values[0], data.loc[i].values[3]], [0,60]))

pd.DataFrame(Approx_AOC1).describe()
```

```python
for i in range(len(data)):
    if AOC[i] < Approx_AOC1[i]:
        print(f'Case {i}: AOC = {AOC[i]}, Approx_AOC = {Approx_AOC1[i]}')
```

```python
mean_absolute_error(AOC, Approx_AOC1)
```

### Stage 2 AOC Approximation

#### Part 1

FGLU => GLU15 => GLU30<br>
GLU60 => GLU90 => GLU120

```python
glu15_preds_f = glu15_from_fglu(fglu_values)
glu30_preds_f = glu30_from_glu15(glu15_preds_f)
glu90_preds_f = glu90_from_glu60(glu60_values)
glu120_preds_f = glu120_from_glu90(glu90_preds_f)
```

```python
Approx_AOC2_1 = []

for i in range(len(data)):
    Approx_AOC2_1.append(np.trapz([data.loc[i].values[0], glu15_preds_f[i], glu30_preds_f[i], data.loc[i].values[3], glu90_preds_f[i],
                                 glu120_preds_f[i]], x))
    
pd.DataFrame(Approx_AOC2_1).describe()
```

```python
mean_absolute_error(Approx_AOC2_1, AOC)
```

#### Part 2

GLU15 <= GLU30 <= GLU60 => GLU90 => GLU120

```python
glu30_preds_b = glu30_from_glu60(glu60_values)
glu15_preds_b = glu15_from_glu30(glu30_preds_b)
```

```python
Approx_AOC2_2 = []

for i in range(len(data)):
    Approx_AOC2_2.append(np.trapz([data.loc[i].values[0], glu15_preds_b[i], glu30_preds_b[i], data.loc[i].values[3], glu90_preds_f[i],
                                 glu120_preds_f[i]], x))
    
pd.DataFrame(Approx_AOC2_2).describe()
```

```python
mean_absolute_error(Approx_AOC2_2, AOC)
```

#### Part 3

FGLU => GLU15 => GLU30 => (GLU60) => GLU90 => GLU120

```python
glu60_preds_fake = glu60_from_glu30(glu30_preds_f)
glu90_preds_f1 = glu90_from_glu60(glu60_preds_fake)
glu120_preds_f1 = glu120_from_glu90(glu90_preds_f1)
```

```python
Approx_AOC2_3 = []

for i in range(len(data)):
    Approx_AOC2_3.append(np.trapz([data.loc[i].values[0], glu15_preds_f[i], glu30_preds_f[i], data.loc[i].values[3], glu90_preds_f1[i],
                                 glu120_preds_f1[i]], x))
    
pd.DataFrame(Approx_AOC2_3).describe()
```

```python
mean_absolute_error(Approx_AOC2_3, AOC)
```

#### Part 4

GLU15 <= GLU30 <= GLU60 => GLU90 => GLU120<br>
FGLU => GLU15 => GLU30

```python
glu90_preds = glu90_preds_f
glu120_preds = glu120_preds_f
glu30_preds = (glu30_preds_f*0.9*0.95 + glu30_preds_b*0.94)/(0.9*0.95+0.94)
glu15_preds = (glu15_preds_f*0.95 + glu15_preds_b*0.95*0.92)/(0.95+0.95*0.92)
```

```python
Approx_AOC2_4 = []

for i in range(len(data)):
    Approx_AOC2_4.append(np.trapz([data.loc[i].values[0], glu15_preds[i], glu30_preds[i], data.loc[i].values[3], glu90_preds[i],
                                 glu120_preds[i]], x))
    
pd.DataFrame(Approx_AOC2_4).describe()
```

```python
mean_absolute_error(Approx_AOC2_4, AOC)
```

We observe that Part 1 performs as good as Part 4, that is GLU60 alone can do good enough predictions, introducing FGLU actually worsens predictions?

### Approach 2

```python
fglu_values = np.array(data['FGLU'].values)
glu15_values = np.array(data['GLU15'].values)
glu30_values = np.array(data['GLU30'].values)
glu60_values = np.array(data['GLU60'].values)
glu90_values = np.array(data['GLU90'].values)
glu120_values = np.array(data['GLU120'].values)
```

```python
np.random.seed(0)

def obtain_linear_regressor(inputs, output, train_size):
    model = LinearRegression()
    X = np.column_stack(inputs)
    y = np.array(output)
    X_train, X_val, y_train, y_val = train_test_split(X, y, train_size = train_size, random_state=10)
    model.fit(X_train, y_train)
    
    train_preds = model.predict(X_train)
    print('On Training Data')
    print('Score:',model.score(X_train, y_train))
    print('MAE:', mean_absolute_error(train_preds, y_train))

    val_preds = model.predict(X_val)
    print('On Validation Data')
    print('Score:',model.score(X_val, y_val))
    print('MAE:', mean_absolute_error(val_preds, y_val))
    
    return model
```

#### Part 1

(FGLU, GLU60) => GLU15 <br>
(FGLU, GLU60, GLU15) => GLU30<br>
(FGLU, GLU60, GLU15, GLU30) => GLU90<br>
(FGLU, GLU60, GLU15, GLU30, GLU90) => GLU120<br>

```python
lr1 = obtain_linear_regressor([fglu_values, glu60_values], glu15_values, 0.8)
lr2 = obtain_linear_regressor([fglu_values, glu60_values, glu15_values], glu30_values, 0.8)
lr3 = obtain_linear_regressor([fglu_values, glu60_values, glu15_values,glu30_values], glu90_values, 0.8)
lr4 = obtain_linear_regressor([fglu_values, glu60_values, glu15_values,glu30_values, glu90_values], glu120_values, 0.8)
```

```python
glu15_preds3_1 = lr1.predict(np.column_stack((fglu_values, glu60_values)))
glu30_preds3_1 = lr2.predict(np.column_stack((fglu_values, glu60_values, glu15_preds3_1)))
glu90_preds3_1 = lr3.predict(np.column_stack((fglu_values, glu60_values, glu15_preds3_1, glu30_preds3_1)))
glu120_preds3_1 = lr4.predict(np.column_stack((fglu_values, glu60_values, glu15_preds3_1, glu30_preds3_1, glu90_preds3_1)))
```

```python
Approx_AOC_Approach3_1 = []
for i in range(len(data)):
        Approx_AOC_Approach3_1.append(np.trapz([data.loc[i].values[0], glu15_preds3_1[i], glu30_preds3_1[i],data.loc[i].values[3],
                                glu90_preds3_1[i], glu120_preds3_1[i]], x))
        
pd.DataFrame(Approx_AOC_Approach3_1).describe()
```
